```python
import pandas as pd
```

```python
df = pd.read_csv("student_performance.csv")
print(df)
```

```python
agg_data = df.groupby("Department").agg({
    "Marks": ["sum", "mean", "max"],
    "Attendance": "mean"
})
print(agg_data)
```

```python
df["Marks_Rank"] = df.groupby("Department")["Marks"].transform(
    lambda x: x.rank(ascending=False)
)
print(df)
```

```python
filtered_df = df.groupby("Department").filter(
    lambda x: x["Marks"].mean() > 85
)
print(filtered_df)
```

```python
scholarship = pd.DataFrame({
    "Student_ID": [101, 102, 103, 104, 105],
    "Scholarship": [5000, 4000, 6000, 3500, 4500]
})
merged_df = pd.merge(
    df,
    scholarship,
    on="Student_ID",
    how="inner"
)
print(merged_df)
```

```python
left_merge = pd.merge(
    df,
    scholarship,
    on="Student_ID",
    how="left"
)
print(left_merge)
```

```python
pivot_marks = df.pivot_table(
    index="Department",
    columns="Gender",
    values="Marks",
    aggfunc="mean"
)
print(pivot_marks)
```

```python
pivot_attendance = df.pivot_table(
    index="Department",
    columns="Gender",
    values="Attendance",
    aggfunc="mean"
)
print(pivot_attendance)
```

```python
pivot_marks.to_csv("department_marks_report.csv")
```
