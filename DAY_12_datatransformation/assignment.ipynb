```python
import pandas as pd
import numpy as np
```

```python
# Create Employee Dataset
data = {
    'employee_id': [101, 102, 103, 104, 105],
    'employee_name': ['Akhil', 'Meghana', 'Rohit', 'Pooja', 'Harish'],
    'department': ['IT', 'HR', 'Finance', 'Marketing', 'IT'],
    'salary': [50000, 45000, 60000, 55000, 70000],
    'bonus': [5000, 4000, 6000, 5500, 7000],
    'joining_date': [
        '2021-01-15',
        '2020-06-20',
        '2019-03-10',
        '2022-08-05',
        '2018-11-01'
    ]
}
```

```python
df = pd.DataFrame(data)
```

```python
# Display Dataset
print("Original Dataset")
print(df)
```

```python
# Calculate Total Salary
df['total_salary'] = df['salary'] + df['bonus']

df
```

```python
# Convert joining_date to datetime
df['joining_date'] = pd.to_datetime(df['joining_date'])

df
```

```python
# Extract Year
df['joining_year'] = df['joining_date'].dt.year

df
```

```python
# Create Salary Category
df['salary_category'] = np.where(
    df['salary'] >= 55000,
    'High',
    'Medium'
)

df
```

```python
Q1 = df['salary'].quantile(0.25)
Q3 = df['salary'].quantile(0.75)

IQR = Q3 - Q1

outliers = df[
    (df['salary'] < (Q1 - 1.5 * IQR)) |
    (df['salary'] > (Q3 + 1.5 * IQR))
]

print("Outliers:")
print(outliers)
```

```python
print("Transformed Dataset")
print(df)
```
