### Employee Salary Analysis using Seaborn

```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Employee Dataset
data = {
    'Employee': ['ARJUN', 'SNEHA', 'RAVI', 'DIVYA', 'KARTHIK'],
    'Experience': [2, 5, 3, 7, 6],
    'Salary': [35000, 60000, 42000, 75000, 68000]
}

df = pd.DataFrame(data)

# Line Plot
sns.lineplot(x='Employee', y='Salary', data=df, marker='o')
plt.title('Employee Salary Trend')
plt.xlabel('Employee')
plt.ylabel('Salary')
plt.show()

# Scatter Plot
sns.scatterplot(x='Experience', y='Salary', data=df, s=100)
plt.title('Experience vs Salary')
plt.xlabel('Experience (Years)')
plt.ylabel('Salary')
plt.show()

# Box Plot
sns.boxplot(y='Salary', data=df)
plt.title('Salary Distribution')
plt.show()

# Violin Plot
sns.violinplot(y='Salary', data=df)
plt.title('Salary Distribution (Violin Plot)')
plt.show()

# Swarm Plot
sns.swarmplot(x='Employee', y='Salary', data=df, size=8)
plt.title('Employee Salary Swarm Plot')
plt.show()

# Bar Plot
sns.barplot(x='Employee', y='Salary', data=df)
plt.title('Employee Salary Comparison')
plt.show()

# Point Plot
sns.pointplot(x='Employee', y='Experience', data=df)
plt.title('Employee Experience Point Plot')
plt.show()

# Count Plot Dataset
data2 = {
    'Department': [
        'HR', 'IT', 'Finance',
        'IT', 'HR', 'Marketing',
        'IT', 'Finance', 'HR'
    ]
}

df2 = pd.DataFrame(data2)

# Count Plot
sns.countplot(x='Department', data=df2)
plt.title('Department Employee Count')
plt.show()

# Load Tips Dataset
tips = sns.load_dataset('tips')

# KDE Plot
sns.kdeplot(
    data=tips,
    x='total_bill',
    fill=True,
    label='Total Bill'
)
plt.legend()
plt.title('KDE Plot of Total Bill')
plt.show()

# Scatter Plot using Tips Dataset
sns.scatterplot(
    x='total_bill',
    y='tip',
    data=tips
)
plt.title('Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()

# Built-in Style-
sns.set_style("whitegrid")
sns.boxplot(
    x='day',
    y='total_bill',
    data=tips
)
plt.title('Total Bill Distribution by Day')
plt.show()

# Built-in Color Palette
sns.set_palette("pastel")
sns.violinplot(
    x='day',
    y='tip',
    data=tips
)
plt.title('Tip Distribution by Day')
plt.show()

# Custom Color Palette
custom_colors = ['#6A0572', '#AB83A1', '#F0A6CA', '#FFD6E0']
"#ab8395""#FF0073""#AB83A1"
sns.set_palette(custom_colors)
sns.violinplot(
    x='time',
    y='total_bill',
    data=tips
)

plt.title('Custom Colored Total Bill Distribution')
plt.show()

# Conclusion
print("Employee Salary Analysis using Seaborn Completed Successfully.")
```
