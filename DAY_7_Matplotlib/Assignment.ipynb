### Monthly Rainfall Analysis

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Line Chart
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
rainfall = [45, 60, 80, 120, 150, 100]
plt.plot(months, rainfall,
         color='blue',
         marker='o',
         linestyle='--')
plt.title("Monthly Rainfall Trend")
plt.xlabel("Months")
plt.ylabel("Rainfall (mm)")
plt.show()
```

```python
# Bar Chart
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
rainfall = [45, 60, 80, 120, 150, 100]
plt.bar(months, rainfall, color='skyblue')
plt.title("Monthly Rainfall Comparison")
plt.xlabel("Months")
plt.ylabel("Rainfall (mm)")
plt.show()
```

```python

# Histogram
data = [5, 7, 8, 10, 12, 12, 14, 15,
        18, 19, 20, 21, 22, 24, 25, 27]
plt.hist(data, bins=5, color='pink')
plt.title("Histogram Distribution")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

```python
# Scatter Plot

temperature = [22, 24, 26, 28, 30, 32]
rainfall = [45, 60, 80, 120, 150, 100]
plt.scatter(temperature,
            rainfall,
            color='green')
plt.title("Rainfall vs Temperature")
plt.xlabel("Temperature (°C)")
plt.ylabel("Rainfall (mm)")
plt.show()
```

```python
# Pie Chart
usage = ['Domestic', 'Agriculture', 'Industry', 'Others']
percentage = [40, 35, 15, 10]
plt.pie(percentage,
        labels=usage,
        autopct='%1.1f%%')
plt.title("Water Usage Distribution")
plt.show()
```

```python
# Box Plot

rainfall_data = [
    [40, 45, 50, 55, 60],
    [70, 75, 80, 85, 90],
    [100, 110, 120, 130, 140]
]
plt.boxplot(rainfall_data)
plt.title("Rainfall Box Plot")
plt.xlabel("Regions")
plt.ylabel("Rainfall (mm)")
plt.show()
```

```python
# Heatmap
np.random.seed(5)
data = np.random.rand(6, 6)
plt.imshow(data, cmap='coolwarm')
plt.colorbar()
plt.title("Weekly Temperature Heatmap")
plt.show()
```
